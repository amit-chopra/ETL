```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
from config import dbpasswd
import pymysql
pymysql.install_as_MySQLdb()
```

```python
csvfile = "sdf13_1a.csv"
school_df = pd.read_csv(csvfile)
```

```python
school_df.describe()
```

```python
#create a new dataframe with only the columns needed. These columns match those in the SQL DB.
school_new_df = school_df[["FIPST","STNAME","STABBR","YEAR","MEMBERSCH",
                           "TOTALREV","TFEDREV","TSTREV","TLOCREV","TOTALEXP",
                           "TCURELSC","TCURINST","TCURSSVC","TCUROTH","TNONELSE",
                           "TCAPOUT"]].copy()
school_new_df.head()
```

```python
#group school districts by state.
state_dist_group = school_new_df.groupby("FIPST")
```

```python
#calculate the number of school districts in each state
state_dist_count = state_dist_group["FIPST"].count()
state_dist_count.head()
```

```python
#Make the count of state districts
state_districts = pd.DataFrame(state_dist_count)
state_districts.head()
```

```python
#rename the column for district count. Matches SQL database.
state_districts.rename(columns={"FIPST":"DISTCNT"}, inplace=True)
state_districts.head()
```

```python
#Clean up the data.  Schools with no values have a -1 or -2 instead of a zero or NULL. This messes up the math.
school_clean = school_new_df.replace(-1,0)
school_clean.head()
```

```python
#Clean up the data.  Schools with no values have a -1 or -2 instead of a zero or NULL. This messes up the math.
school_clean.replace((-2,0), inplace = True)
school_clean.head()
```

```python
#get the sum of all of the fields that need this information.
schoolstate_rev = school_clean.groupby(["FIPST"])["MEMBERSCH","TOTALREV","TFEDREV","TSTREV","TLOCREV","TOTALEXP",
                           "TCURELSC","TCURINST","TCURSSVC","TCUROTH","TNONELSE",
                           "TCAPOUT"].sum()
schoolstate_rev.head()
```

```python
#All values are the same - getting max so I can get only 1 value per state.
schoolstate_data = school_clean.groupby(["FIPST"])["STNAME", "STABBR", "YEAR"].max()
schoolstate_data.head()
```

```python
#merge the summarized data with the regular state date with the count of districts.
schoolfund_state_df = pd.merge(pd.merge(schoolstate_data, schoolstate_rev, on="FIPST"), state_districts, on="FIPST")
schoolfund_state_df.head()
```

```python
#what do we have now?
schoolfund_state_df.describe()
```

### Connect to the database and load to SQL...

```python
#Setup the connection to SQL
engine = create_engine(f'mysql://root:{dbpasswd}@localhost/Schools')
```

```python
#confirm the tables created in SQL are there.
engine.table_names()
```

```python
#put the data into SQL
schoolfund_state_df.to_sql(name='school_funding', con=engine, if_exists = 'append', index = True )
```

```python
#confirm we have the table
pd.read_sql_query('select * from school_funding', con=engine).head()
```
